```python
import seaborn as sns
```

```python
%ls datasets/
```

```python
df =pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x= 'Wingspan',
           y= 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.show()
```

```python
sns.swarmplot(x= 'Gender',
           y= 'CWDistance',
           data = df
          )
plt.show()
```

# what is the minimum and maximum of heigth and all others

```python
sns.boxplot(data = df.loc[:,
                          ['Height','Wingspan','CWDistance','CompleteGroup','Score']
                         ]
           )
```

```python
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']]
           )
```

```python
fig = plt.figure()
ax1=fig.add_subplot(1,2,1) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']]
           )

fig = plt.figure()
ax2=fig.add_subplot(1,2,2) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']]
           )
ax1.title.set_text('male scores')
ax2.title.set_text('Female scores')
```

```python
fig = plt.figure()
ax1=fig.add_subplot(2,2,1) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']]
           )

fig = plt.figure()
ax2=fig.add_subplot(2,2,2) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']]
           )
ax1.title.set_text('male scores')
ax2.title.set_text('Female scores')
fig = plt.figure()
ax3=fig.add_subplot(2,2,3) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']]
           )

fig = plt.figure()
ax4=fig.add_subplot(2,2,4) #number of rows columns position id
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']]
           )
ax3.title.set_text('male scores')
ax4.title.set_text('Female scores')
```
