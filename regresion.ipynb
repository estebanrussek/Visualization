## Existe una correlaci√≥n entre Head Size y Brain Weight?

```python
df = pd.read_csv('datasets/headbrain.csv')
df.head


X = df['Head Size(cm^3)']
Y = df['Brain Weight(grams)']

np.corrcoef(X,Y)
```

```python
colors = {1: 'red', 2: 'green'}
plt.figure(figsize = (13,8))
plt.scatter(X, Y,
           c = df['Gender'].map(colors),
           marker = 'x',
           s = 175)
plt.xlabel('Head Size(cm^3)', fontweight = 'regular', color = 'black', fontsize = 18, horizontalalignment = 'center')
plt.ylabel('Brain Weight(grams)', fontweight = 'regular', color = 'black', fontsize = 18, horizontalalignment = 'center')
```

## CONCLUSION
Por lo tanto, la coeficiente de correlacion es 0.79956971, podemos decir, existe una relacion muy fuerte entre Head Size y Brain Weight
